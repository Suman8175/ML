```python
import pandas as pd
import requests
```

```python
#hit the api using requests.get()
data=requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page=1')
```

```python
df=pd.DataFrame(data.json()['results'])[['title','popularity','release_date','vote_average','vote_count']]
```

```python
df
```

```python
#as we know how to hit api we have 490 total pages.so hit the api in loop
df=pd.DataFrame()
for i in range(1,491):
    temp_Data=requests.get(f"https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={i}")
    temp_Df=pd.DataFrame(temp_Data.json()['results'])
    temp_Df=temp_Df.loc[:,['title','popularity','release_date','vote_average','vote_count']]
    df=pd.concat([df,temp_Df],ignore_index=True)
```

```python
#exporting the df to csv
df.to_csv('movv.csv', index=False)
```
