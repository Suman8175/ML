# 1. Relational Plot
- to see the statistical relation between 2 or more variables.
- Bivariate Analysis</br>

**Plots under this section**
- scatterplot
- lineplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
tip=sns.load_dataset('tips') #some dummy datasets are given by seaborn 
tip
```

```python
# scatterplot -> axis-level-function
sns.scatterplot(data=tip,x='total_bill',y='tip',color='red')
plt.axhline(y=6)
plt.axvline(x=30)
```

```python
#to use it as figure level function
sns.relplot(data=tip,x='total_bill',y='tip',color='red',kind='scatter')
#prefer it as .relplot() is recommended by documentation and easier as .relplot() is used for all scatter,line,etc graph
```

```python
#now suppose I want to know which dot is which Male or Female
sns.relplot(data=tip,x='total_bill',y='tip',color='red',hue='sex',kind='scatter')
```

```python
#now I want to know which person come at dinner or lunch use style paramter
#also size parameter for size,more big less small
sns.relplot(data=tip,x='total_bill',y='tip',color='red',hue='sex',style='time',size='size',kind='scatter')
```

# Line 

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
tempdata = penguins[penguins['species'].isin(['Adelie','Gentoo'])]
tempdata=tempdata.dropna()
tempdata
```

```python
sns.relplot(data=tempdata,x='body_mass_g',y='bill_length_mm',hue='island',kind='line')
```

```python
#show all the data in different different graph
sns.relplot(data=tempdata,x='body_mass_g',y='bill_length_mm',col='island',kind='line')
```

# 2.Distribution Plot
- used for univariate analysis
- used to find out the distribution
- Range of the observation
- Central Tendency
- is the data bimodal?
- Are there outliers?</br>

**Plots under distribution plot**
- histplot
- kdeplot
- rugplot

```python
# figure level -> displot
# axes level -> histplot -> kdeplot -> rugplot
```

```python
sns.displot(data=tip,x='total_bill',kind='hist',col='sex')
```

# KDE Plot

```python
sns.displot(data=tip,x='total_bill',kind='kde',hue='sex',fill=True)
```

```python
#bivariante plot(both x and y axis have value from data)
sns.displot(data=tip,x='total_bill',y='tip',kind='hist')
#it shows darker area have more value meaning
# whose bill is 15 and tips is 3 have more value than other
```

# 3.Categorical Data
- use to represent category values

# Stripplot

```python
sns.catplot(data=tip,x='day',kind='strip',y='total_bill',palette='viridis',hue='sex')
```

```python
sns.catplot(data=tip,hue='day',kind='box',y='total_bill',palette='viridis')
```

# Bar Plot

```python
#estimator default value is mean,you can change to any you like like std,nanmedian,mean,etc
sns.catplot(data=tip,x='sex',y='total_bill',kind='bar',hue='smoker',estimator=np.nanmedian)
```

# Count Plot
- counts and plot

```python
sns.catplot(data=tip,x='sex',hue='day',kind='count')
```
