**Opening csv from URL**

```python
import requests
from io import StringIO
#url and headers can be different others are same
url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data)
```

**Read TSV file(Tab Separated file)  with no headers**

```python
#names paramter is to add header/column names for each
df=pd.read_csv('data/movies.tsv',sep='\t',names=['sno','names','relesase year','rating','no of ratings','genree'])
df
```

**Skip bad lines**</br>
Suppose our row should have only 5 data but some row may have 7 or 4 or some other value.</br>
In this case it will show parser error expected xxx fields in row no got yyy.

```python
pd.read_csv('data/placement.csv',on_bad_lines='skip')
```
