```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/train.csv')
df.sample(5)
```

```python
#lets count each values
sns.catplot(x=df['Survived'],hue=df['Sex'],kind='count')
```

```python
#count no of people in each class
sns.catplot(x=df['Pclass'],kind='count',hue=df['Sex'])
```

```python
#count how many people 
sns.catplot(x=df['Embarked'],kind='count',hue=df['Sex'])
```

```python
rr=df['Pclass'].value_counts()
plt.pie(rr,labels=rr.index,autopct='%1.0f%%')
```
