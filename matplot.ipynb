```python
df=pd.read_csv('data/sharma-kohli.csv')
```

```python
df
```

```python
plt.plot(df['index'],df['V Kohli'])
plt.plot(df['index'],df['RG Sharma'])
```

Now it might be confusing like what this graph represent?</br>
What does x-axis represents?</br>
What does y-axis represents?</br>
to solve it use <mark>.title()</mark> ,<mark> .xlabel()</mark> , <mark>.ylabel()</mark>

```python
plt.plot(df['index'],df['V Kohli'])
plt.plot(df['index'],df['RG Sharma'])
plt.title('Career comparision of Virat Kolhi vs Rohit Sharma')
plt.xlabel('Years')
plt.ylabel('No of runs scored')
```

Now we might want to give red color to virat and green color to rohit.</br>
pass <mark> color </mark> paramter 

```python
plt.plot(df['index'],df['V Kohli'],color='red')
plt.plot(df['index'],df['RG Sharma'],color='green')
plt.title('Career comparision of Virat Kolhi vs Rohit Sharma')
plt.xlabel('Years')
plt.ylabel('No of runs scored')
```

We can also use different lines like dash,dot,etc</br>
pass <mark>linestyle</mark> parameter

```python
plt.plot(df['index'],df['V Kohli'],color='red',linestyle='dashed')
plt.plot(df['index'],df['RG Sharma'],color='green',linestyle='dotted')
plt.title('Career comparision of Virat Kolhi vs Rohit Sharma')
plt.xlabel('Years')
plt.ylabel('No of runs scored')
```

```python
# legend (labels to show which line represents which)
plt.plot(df['index'],df['V Kohli'],color='red',linestyle='dashed',marker='o',label='Virat Kolhi')
plt.plot(df['index'],df['RG Sharma'],color='green',linestyle='dotted',label='Rohit Sharma')
plt.title('Career comparision of Virat Kolhi vs Rohit Sharma')
plt.xlabel('Years')
plt.ylabel('No of runs scored')
plt.legend()
plt.show()
```

```python
#now we might only want to see between 2012 to 2015- use plt.ylim() for y limit and plt.xlim for x limit
plt.plot(df['index'],df['V Kohli'],color='red',linestyle='dashed',marker='o',label='Virat Kolhi')
plt.plot(df['index'],df['RG Sharma'],color='green',linestyle='dotted',label='Rohit Sharma')
plt.title('Career comparision of Virat Kolhi vs Rohit Sharma 2012-2015')
plt.xlabel('Years')
plt.ylabel('No of runs scored')
plt.legend()
plt.xlim(2012,2015)
plt.ylim(300,750)
```

```python
#to turn on grid use plt.grid()
plt.plot(df['index'],df['V Kohli'],color='red',linestyle='dashed',marker='o',label='Virat Kolhi')
plt.plot(df['index'],df['RG Sharma'],color='green',linestyle='dotted',label='Rohit Sharma')
plt.title('Career comparision of Virat Kolhi vs Rohit Sharma')
plt.xlabel('Years')
plt.ylabel('No of runs scored')
plt.legend()
plt.grid()
```

# Bar-Graph
- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

```python
x=['Mon','Tue','Wed','Thu','Fri']
y=[10,30,25,10,20]
colors=['red','blue','black','green','pink']
plt.bar(x,y,color=colors)
plt.title('Students present as per day')
plt.xlabel('Days')
plt.ylabel('No of students')
```

```python
df = pd.read_csv('data/batsman_season_record.csv')
df
```

```python
#stacked bargraph
plt.bar(df['batsman'],df['2015'],label='2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],label='2016')
plt.bar(df['batsman'],df['2017'],bottom=df['2015']+df['2016'],label='2017')
plt.legend()
```

# PieCharts

```python
marks=np.random.randint(30,105,5)
label=['Sun','Mon','Tue','Wed','Thu']
plt.pie(marks,labels=label,autopct='%1.0f%%',explode =[0.1,0,0,0,0])
#explode to take out certain according to insert position 0.1 position out to 1st rest 0
```
