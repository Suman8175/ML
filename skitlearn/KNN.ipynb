```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv('../data/breast-cancer.csv',index_col=False)
df=df.drop(columns=['id','Unnamed: 32'])
df.sample(10)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df.iloc[:,1:], df.iloc[:,0], test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_Scalar_train=scaler.transform(X_train)
X_Scalar_test=scaler.transform(X_test)
```

```python
X_Scalar_train=pd.DataFrame(X_Scalar_train)
X_Scalar_train.sample(10)
```

```python
#knn
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=5)
neigh.fit(X_Scalar_train, y_train)
```

```python
#predict

from sklearn.metrics import accuracy_score
y_pred=neigh.predict(X_Scalar_test)
accuracy_score(y_test, y_pred)
```
